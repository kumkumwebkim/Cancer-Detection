```python
# dependent variable and independent variable

x=data[['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

y=data['diagnosis']
```

```python
# train and test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# what KNN means?
#K= the number of nearest neighbors to consider for classification.
# N = the number of features in the dataset(in this case, 30).
```

```python
# import model
from sklearn.neighbors import KNeighborsClassifier

knn= KNeighborsClassifier(n_neighbors=3)
```

```python
x_train_numeric = x_train.drop(['id', 'diagnosis'], axis=1)
knn.fit(x_train_numeric, y_train)
```

```python
# make prediction
x_test_numeric = x_test.drop(['id', 'diagnosis'], axis=1)
y_pred=knn.predict(x_test_numeric)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of KNN model: {accuracy * 100:.2f}%')
```

# Confusion matrix


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm= confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
cm_display.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for KNN Model')
plt.show()
```

```python
# precision, recall, f1- score for KNN model
from sklearn.metrics import precision_score, recall_score, f1_score

from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred, target_names=['Benign', 'Malignant'])
print("Classification Report:")
print(report)
# ROC curve and AUC
from sklearn.metrics import roc_curve, auc

# Convert 'diagnosis' column to numerical values
y_test_numeric = y_test.map({'M': 1, 'B': 0})

fpr, tpr, thresholds = roc_curve(y_test_numeric, knn.predict_proba(x_test_numeric)[:, 1])
roc_auc = auc(fpr, tpr)
```
